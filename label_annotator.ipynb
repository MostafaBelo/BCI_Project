```python
import csv
```

```python
# -------- STEP 1: Load CSV and map sentence -> sentiment --------
csv_file = "/mnt/D/University/Fall 2025/BCI/Dataset/sentiment_labels_task1.csv"
sentence_to_label = {}

with open(csv_file, encoding="utf-8") as f:
    reader = csv.DictReader(f, delimiter=";")
    for row in reader:
        try:
            sentence = row["sentence"].strip()
            label = row["sentiment_label"].strip()
            sentence_to_label[sentence] = label
        except:
            pass


# -------- STEP 2: Read the ordered sentences --------
input_file = "shards/labels_original.txt"
output_file = "shards/labels_sentiment.txt"

results = []

with open(input_file, encoding="utf-8") as f:
    for line in f:
        # split: "0:Sentence..."
        idx, sentence = line.split(":", 1)
        sentence = sentence.strip()

        # find matching sentiment
        label = sentence_to_label.get(sentence, "NOT_FOUND")

        results.append(f"{idx}:{label}:{sentence}")


# -------- STEP 3: Write new file --------
with open(output_file, "w", encoding="utf-8") as f:
    for r in results:
        f.write(r + "\n")

print("Merged file written to:", output_file)
```

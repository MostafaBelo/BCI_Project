```python
import localizer
import numpy as np
from tqdm import tqdm
```

```python
eeg_signal = mat["sentenceData"]["rawData"][0,2]

localized_eeg = localizer.localize(eeg_signal)

localized_eeg.shape, eeg_signal.shape
```

```python
res = np.empty((1, 400), dtype=object)
```

```python
for i in tqdm(range(400)):
    eeg_signal = mat["sentenceData"]["rawData"][0,i]
    try:
        localized_eeg = localizer.localize(eeg_signal)
        res[0,i] = localized_eeg
    except Exception as e:
        print(f"Error processing sentence {i}: {e}")
        print(eeg_signal.shape)
        res[0,i] = None
```

```python
sum(x.nbytes for x in res[0] if x is not None) / (1024**3)  # in GB
```
